```python
x = np.arange(0,10,0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
x = np.linspace(-4,4,100)
y1 = x**3
y2 = 10*x**2 - 2

plt.plot(x, y1, x, y2)
plt.show()
```

```python
plt.plot(x,y1,x, y2)
plt.xlim(-1, 1)
plt.ylim(-3, 3)
plt.show()
```

```python
x = np.arange(0,5,1)
y1 = x
y2 = x+1
y3 = x+2
y4 = x+3
```

```python
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
plt.plot(x, y1, '-', x, y2, '--', x, y3, ':', x, y4, '-.')
plt.show()
```

```python
city = ['', '', '', '', '', '', '']

lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07,126.85]

pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]


size = np.array(pop_den)
colors = ['r','g','b', 'c', 'm', 'k' ,'y']

plt.scatter(lon, lat, s=size, c=colors, alpha = 0.5)
plt.xlabel('(longitude)')
plt.ylabel('(latitude)')
plt.title('  (2017)')

for x,y, name in zip(lon, lat, city):
    plt.text(x,y,name)

plt.show()
```
