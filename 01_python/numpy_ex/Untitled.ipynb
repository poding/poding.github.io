```python
x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
plt.plot(x, y1, x, y2, x, y3)  # 한 번의 호출로 표현
plt.show()
```

```python
plt.plot(x, y1)
plt.figure()  # 새로운 그래프 창 생성
plt.plot(x, y2)
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2

plt.figure(1)  # 1번 그래프 창
plt.plot(x, y1)

plt.figure(2)  # 2번 그래프 창
plt.plot(x, y2)

plt.figure(1)  # 1번 그래프에 추가
plt.plot(x, y2)

plt.figure(2)  # 2번 그래프에 추가
plt.clf()  # 2번 그래프 모두 지우기
plt.plot(x, y1)

plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1

# 2x2 행렬 그래프
plt.subplot(2,2,1)  # 1번째 그래프 지정
plt.plot(x, y1)

plt.subplot(2,2,2)  # 2번째 그래프 지정
plt.plot(x, y2)

plt.subplot(2,2,3)  # 3번째 그래프 지정
plt.plot(x, y3)

plt.subplot(2,2,4)  # 4번째 그래프 지정
plt.plot(x, y4)

plt.show()
```

```python
x = np.linspace(-4, 4, 100)  # [-4, 4] 범위에서 100개의 값 생성
y1 = x**3
y2 = 10*x**2 - 2

plt.plot(x, y1, x, y2)
plt.show()
```

```python
plt.plot(x, y1, x, y2)
plt.xlim(-1, 1)  # 출력 범위 지정(확대해서 출력)
plt.ylim(-3, 3)
plt.show()
```

# 그래프 꾸미기

```python
x = np.arange(0, 5, 1)  # 색상 지정
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
```

```python
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
plt.plot(x, y1, '-', x, y2, '--', x, y3, ':', x, y4, '-.')  # 선 모양 변경
plt.show()
```

```python
plt.plot(x, y1, 'o', x, y2, '^', x, y3, 's', x, y4, 'd')  # 점 모양 변경
plt.show()
```

```python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')  # 선, 점 합쳐서 표기
plt.show()
```

```python
x = np.arange(-4.5, 5, 0.5)
y = 2*x**3

plt.plot(x,y)
plt.xlabel('X-axis')  # 가로, 세로 축 라벨링
plt.ylabel('Y-axis')
plt.title('Graph title')  # 제목
plt.grid(True)  # plt.grid()도 가능
plt.show()
```

```python
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.legend(['data1', 'data2', 'data3', 'data4'])
plt.show()
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # 맑은 고딕으로 설정
matplotlib.rcParams['axes.unicode_minus'] = False

plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.legend(['데이터1', '데이터2', '데이터3', '데이터4'], loc = 'best')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('그래프 제목')
plt.grid(True)
```

```python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.text(0, 6, "문자열 출력 1")  # 도표의 x, y 좌표
plt.text(0, 5, "문자열 출력 2")
plt.text(3, 1, "문자열 출력 3")
plt.text(3, 0, "문자열 출력 4")
plt.show()
```

# 산점도

```python
# 기본값 : s=40, c='b', marker='o', alpha=1
```

```python
height = [165, 177, 160, 180, 185, 155, 172]
weight = [62, 67, 55, 74, 90, 43, 64]

plt.scatter(height, weight)
plt.xlabel('Height(m)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')
plt.grid(True)
```

```python
plt.scatter(height, weight, s=500, c='r')  # 크기:500, 색상:Red
plt.show()
```

```python
size = 100 * np.arange(1, 8)  # 데이터 별로 크기 지정
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']  # 데이터 별로 색상 지정

plt.scatter(height, weight, s=size, c=colors)
plt.show()
```

```python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

# 위도와 경도
lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

# 인구 밀도
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den) * 0.2  # 크기 지정
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']  # 색상 지정

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017)')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name)  # 위도, 경도에 맞게 도시 이름 출력
    
plt.show()
```
